## Seperate X and Y

```python
X = df.drop(columns=['PROFIT'])
Y = df[['PROFIT']]
```

```python
X.head()
```

```python
Y.head()
```

## Create a Preprocessing Pipeline for X

```python
cat = list(X.columns[X.dtypes=='object'])
con = list(X.columns[X.dtypes!='object'])
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.compose import ColumnTransformer 
```

```python
num_pipe = Pipeline(steps=[('impute',SimpleImputer(strategy='mean')),
                           ('scaler',StandardScaler())])
cat_pipe = Pipeline(steps=[('impute',SimpleImputer(strategy='most_frequent')),
                           ('ohe',OneHotEncoder(handle_unknown='ignore'))])
pre = ColumnTransformer([('num',num_pipe,con),('cat',cat_pipe,cat)])
```

```python
X_pre = pre.fit_transform(X)
X_pre[0:5]
```

```python
cols = pre.get_feature_names_out()
cols
```

```python
X_pre = pd.DataFrame(X_pre,columns=cols)
X_pre.head()
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre,Y,test_size=0.2,random_state=42)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
ytrain.head()
```

```python
xtrain.head()
```

```python
xtest.head()
```

### Create Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor(max_depth=1,min_samples_split=11,min_samples_leaf=5,criterion='squared_error')
dtr.fit(xtrain,ytrain)
```

```python
dtr.score(xtrain, ytrain)
```

```python
dtr.score(xtest, ytest)
```

## Apply Gridserchcv for hyperparameter Tuning

```python
params = {'max_depth':[2,3,4,5,6,7,8,9,10],
          'min_samples_split':[5,6,7,8,9],
          'min_samples_leaf':[4,5,6,7,8],
          'criterion':['squared_error','absolute_error']}
```

```python
from sklearn.model_selection import GridSearchCV
dtr = DecisionTreeRegressor()
gscv = GridSearchCV(dtr,param_grid=params,cv=5,scoring='neg_mean_squared_error')
gscv.fit(xtrain, ytrain)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```

```python
best_dtr = gscv.best_estimator_
best_dtr
```

## Evaluate the Model

```python
best_dtr.score(xtrain, ytrain)
```

```python
best_dtr.score(xtest, ytest)
```
