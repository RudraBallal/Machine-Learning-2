## Seperate X and Y

```python
X = df.drop(columns=['species'])
Y = df[['species']]
```

```python
X.head()
```

```python
Y.head()
```

## Perform Preprocessing for X

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
# numerical_pipeline
num_pipe = Pipeline(steps=[('impute',SimpleImputer(strategy='mean')),
                           ('scaler',StandardScaler())])
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre[0:5]
```

```python
cols = num_pipe.get_feature_names_out()
cols
```

```python
X_pre = pd.DataFrame(X_pre,columns=cols)
X_pre.head()
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.33,random_state=42)
```

```python
xtrain.shape
```

```python
ytrain.shape
```

## Apply Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(xtrain, ytrain)
```

## Check model accuracy

```python
model.score(xtrain, ytrain)
```

```python
model.score(xtest, ytest)
```

## Show confusion matrix for test data

```python
ypred_tr = model.predict(xtrain)
ypred_ts = model.predict(xtest)
```

```python
ypred_tr[0:5]
```

```python
ypred_ts[0:5]
```

```python
model.classes_
```

## Estimate the probabilities for each class

```python
yprob_tr = model.predict_proba(xtrain)
yprob_ts = model.predict_proba(xtest)
```

```python
yprob_tr[0:5]
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cf = confusion_matrix(ytest, ypred_ts)
cfd = ConfusionMatrixDisplay(cf, display_labels=model.classes_)
cfd.plot()
```

## Apply classification report on test data

```python
from sklearn.metrics import classification_report
print(classification_report(ytest, ypred_ts))
```
