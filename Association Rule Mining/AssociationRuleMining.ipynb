### Read the Data with open file Method

1. 'r' - read (loading file)
2. 'w' - write (saving a new file)
3. 'a' - modifiying existing file

```python
with open('store_data.csv', 'r') as file1:
    lines = file1.readlines()
```

```python
len(lines)
```

```python
lines[0:5]
```

```python
a = lines[0].strip()
a
```

```python
a.split(',')
```

```python
records = []
for i in lines:
    a = i.strip()
    cart = a.split()
    records.append(cart)
```

```python
len(records)
```

```python
records[0:5]
```

### Deciding minimum support, confidence and lift

```python
sup = (7*5)/len(records)
sup
```

```python
conf = 0.2
```

```python
lift = 3
```

```python
from apyori import apriori
association_rules = apriori(records, min_support=sup, min_conf=0.2, min_lift=3)
association_results = list(association_rules)
```

```python
len(association_results)
```

```python
association_results
```

### Save above in results.txt

```python
with open('Reults.txt', 'w') as file2:
    for i in association_results:
        file2.write(str(i)+'\n\n')
```
